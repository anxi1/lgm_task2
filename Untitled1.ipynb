```python
import pandas as pd
```

```python
df = pd.read_csv("C:\\Users\\sneha\\Documents\\letsgrowmore\\globalterrorismdb_0718dist.csv", encoding='ISO-8859-1')
```

```python
df.head(50)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Count the number of occurrences of each country in the dataset
country_counts = df['country_txt'].value_counts()
```

```python
# Get the top 10 countries with the highest number of terrorist incidents
top_countries = country_counts.head(10)
```

```python
# Plot the top 10 countries affected by terrorism
plt.figure(figsize=(12, 6))
sns.barplot(x=top_countries.index, y=top_countries.values, palette='viridis')
plt.xlabel('Country')
plt.ylabel('Number of Terrorist Incidents')
plt.title('Top 10 Countries Most Affected by Terrorism')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Count the number of occurrences of each region in the dataset
region_counts = df['region'].value_counts()
```

```python
# Get the top 10 regions with the highest number of terrorist incidents
top_regions = region_counts.head(10)
```

```python

# Plot the top 10 regions affected by terrorism
plt.figure(figsize=(10, 6))
sns.barplot(x=top_regions.index, y=top_regions.values, palette='coolwarm')
plt.xlabel('Region')
plt.ylabel('Number of Terrorist Incidents')
plt.title('Top 10 Regions Most Affected by Terrorism')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
